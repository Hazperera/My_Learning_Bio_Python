```python
ncov_dna_record = SeqIO.read("Covid_sequence.fasta.txt","fasta") # Read the sequence record in the file
```

```python
type(ncov_dna_record)
```

```python
ncov_dna_record
```

```python
ncov_dna = ncov_dna_record.seq
ncov_dna # Display the Nucleotides
```

```python
len(ncov_dna) # Length of our sequence
```

```python
ncov_mRNA = ncov_dna.transcribe() # Transcribe (DNA to mRNA)
ncov_mRNA
```

```python
ncov_protein = ncov_mRNA.translate() # Translate to Protein/Amino Acids (mRNA to AA)
ncov_protein
```

```python
len(ncov_protein) # Length of Protein/Amino Acids
```

```python
len(ncov_mRNA)/3 # Check if it is true by dividing it by 3 for codon
```

```python
ncov_amino_acids = ncov_protein.split('*') # Find all the amino acids
# ncov_amino_acids
```

```python
for i in ncov_amino_acids:
    print(i)
```

```python
ncov_aa = [str(i) for i in ncov_amino_acids]
ncov_aa
```

```python
import heapq # Find the largest amino acid
```

```python
heapq.nlargest(10,ncov_aa)
```

```python
for i in ncov_amino_acids:
    if len(i) > 20:
        print(i)
```

```python
import pandas as pd # Place our Amino Acids into a DataFrame
```

```python
df = pd.DataFrame({'amino_acids':ncov_aa})
df.head()
```

```python
df['count'] = df['amino_acids'].apply(len)
df.head()
```

```python
df['count'].nlargest(20) # Find the largest amino acid sequence
```

```python
from Bio.PDB import PDBParser,MMCIFParser
```

```python
parser = PDBParser() # Reading a PDB File
structure = parser.get_structure("mmdb_6LU7", "mmdb_6LU7.txt")
structure
```

```python
model = structure[0] # Chains in the Protein Structure
```

```python
for chain in model:
    print(f'chain {chain},chain_ID: {chain.id}')
```

```python
for model in structure: # Check the atoms
    print(model)
    for chain in model:
        print(chain)
        for residue in chain:
            for atom in residue:
                print(atom)
```
